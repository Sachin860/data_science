```python
regionwisesales=df.groupby('Region')['Sales'].sum()
```

```python
regionwisesales
```

```python
print(type(regionwisesales))  #in series there will no name for column 2 for y_axis
```

```python
rdf=regionwisesales.to_frame(name='sum of sales')
print(rdf)
graph=rdf.reset_index()
print(graph)
```

```python
import matplotlib.pyplot as pit
graph.plot(x='Region',y='sum of sales',kind='bar')
```

```python
regionwisesales2=df.groupby('Region')[['Sales','Profit','Quantity']].sum()
```

```python
regionwisesales2
```

```python
plot=regionwisesales2.reset_index()
plot
```

```python
plot.plot(x='Region',y=['Sales','Profit','Quantity'],kind='bar',color=['red','orange','brown'],subplots=True)
```

```python
plot.plot(x='Region',y=['Sales','Profit','Quantity'],kind='bar',color=['red','orange','brown'],subplots=True,title=['Region vs Sales','Region vs profit','region vs quantity'])
```

```python
plot.plot(x='Region',y=['Sales','Profit','Quantity'],kind='bar',color=['red','orange','brown'],subplots=True,title=['Region vs Sales','Region vs profit','region vs quantity'],legend=False)
```
