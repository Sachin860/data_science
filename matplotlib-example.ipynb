```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700]
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)
# Plot the data
plt.figure(figsize=(15, 5))
plt.plot(df['Month'], df['Sales'], marker='*')

# Add title and labels
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Product A': [150, 200, 250, 300, 350, 400],
    'Product B': [180, 220, 260, 310, 370, 420]
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)

# Create a list of bar positions
bar_width = 0.20
bar_positions = np.arange(len(df['Month']))

# Plot the data
fig, ax = plt.subplots(figsize=(10, 5))

# Create bars for Product A
bar1 = ax.bar(bar_positions, df['Product A'], width=bar_width, label='Product A')

# Create bars for Product B, offset by the width of the bars
bar2 = ax.bar(bar_positions + bar_width, df['Product B'], width=bar_width, label='Product B')

# Add title and labels
ax.set_title('Monthly Sales of Products A and B')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_xticks(bar_positions + bar_width / 2)
ax.set_xticklabels(df['Month'])

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Week 1': [23, 25, 19, 22, 30, 35, 28],
    'Week 2': [28, 30, 21, 25, 32, 38, 33],
    'Week 3': [26, 27, 20, 23, 31, 36, 30],
    'Week 4': [24, 29, 22, 26, 33, 39, 34]
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)
df.set_index('Day', inplace=True)

# Plot the heatmap
plt.figure(figsize=(15, 6))
sns.heatmap(df, annot=True, cmap='BuGn')

# Add title and labels
plt.title('Heatmap of Weekly Data')
plt.xlabel('Week')
plt.ylabel('Day')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Product A': [150, 200, 250, 300, 350],
    'Product B': [180, 220, 260, 310, 370],
    'Product C': [120, 140, 180, 230, 290]
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)

# Plot the stacked bar chart
df.set_index('Month').plot(kind='bar', stacked=True, figsize=(10, 6))

# Add title and labels
plt.title('Monthly Sales of Products A, B, and C')
plt.xlabel('Month')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data1 = {
    'x': np.arange(1, 11),
    'y1': np.random.randint(1, 20, size=10)
}

data2 = {
    'x': np.arange(1, 11),
    'y2': np.random.randint(1, 20, size=10)
}

data3 = {
    'x': np.arange(1, 11),
    'y3': np.random.randint(1, 20, size=10)
}

# Create DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# Create a figure and a grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns

# First subplot
axs[0, 0].plot(df1['x'], df1['y1'], marker='o')
axs[0, 0].set_title('Scatter Plot 1')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y1-axis')

# Second subplot
axs[0, 1].plot(df2['x'], df2['y2'], marker='s', color='r')
axs[0, 1].set_title('Scatter Plot 2')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y2-axis')

# Third subplot
axs[1, 0].plot(df3['x'], df3['y3'], marker='^', color='g')
axs[1, 0].set_title('Scatter Plot 3')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y3-axis')

# Fourth subplot (empty)
axs[1, 1].axis('off')  # Leave this subplot empty

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [10, 20, 30, 25, 15]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(df['Category'], df['Values'], color='skyblue')

# Add title and labels
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Category')

# Show the plot
plt.show()
```
