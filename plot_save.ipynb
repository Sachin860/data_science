```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example dataset
data = sns.load_dataset("iris")

# Define the plot function
def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Create a figure
plt.figure(figsize=(10, 8))

# Apply Seaborn style and add the subplot
with sns.axes_style('darkgrid'):
    plt.subplot(221)
    plot()
    plt.title('Line plot of Sepal Length vs Sepal Width')

# Adding additional subplots for demonstration
with sns.axes_style('darkgrid'):
    plt.subplot(222)
    sns.scatterplot(x="sepal_length", y="sepal_width", data=data)
    sns.despine()
    plt.title('Scatter plot of Sepal Length vs Sepal Width')

with sns.axes_style('darkgrid'):
    plt.subplot(223)
    sns.histplot(data['sepal_length'], kde=False)
    sns.despine()
    plt.title('Histogram of Sepal Length')

with sns.axes_style('darkgrid'):
    plt.subplot(224)
    sns.boxplot(x="species", y="sepal_length", data=data)
    sns.despine()
    plt.title('Box plot of Sepal Length by Species')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Date': ['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04'],
    'Sales': [100, 150, 200, 250]
}

# Create DataFrame
df = pd.DataFrame(data)
print(df)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calculate the running sum of sales
df['Running_Sum'] = df['Sales'].cumsum() #cumlating sum which add ist row and 2nd row =2nd row
print(df)
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sales'], marker='o', label='Daily Sales')
plt.plot(df['Date'], df['Running_Sum'], marker='o', label='Running Sum of Sales', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Daily Sales and Running Sum of Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats

data = {
    "location":['bangalore','chennai','delhi','bangalore','chennai','delhi','bangalore','chennai','delhi','bangalore','chennai','delhi','bangalore','chennai','delhi'],
    'company':['samsung','sony','google','samsung','sony','google','samsung','sony','google','samsung','sony','google','samsung','sony','google'],
    'salary':[36000,45000,44000,31000,42000,51000,50000,16000,38000,15000,18000,25000,52000,22000,42000]
}

df=pd.DataFrame(data)
print(df)

# Perform ANOVA using scipy for comparison
f_statistic_scipy, p_value = stats.f_oneway(
    df[df['location'] == 'chennai']['salary'],
    df[df['location'] == 'delhi']['salary'],
    df[df['location'] == 'bangalore']['salary']
)
    
print(f"F-Statistic (scipy): {f_statistic_scipy}")
print(f"P-Value: {p_value}")
if p_value < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```

```python
f_statistic_scipy, p_value = stats.f_oneway(
    df[df['company'] == 'samsung']['salary'],
    df[df['company'] == 'sony']['salary'],
    df[df['company'] == 'google']['salary']
)
print(f"F-Statistic (scipy): {f_statistic_scipy}")
print(f"P-Value: {p_value}")
if p_value < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```
